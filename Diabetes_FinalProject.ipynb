```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diabetes = pd.read_excel('sample_data/Dataset 1 _ Pima Indians diabetes dataset (PIDD).xlsx')
diabetes.head()
```

```python
plt.figure(figsize=(15, 15))

for idx, feature in enumerate(diabetes.columns) :
    ax = plt.subplot(3, 3, idx + 1)
    sns.histplot(diabetes[feature], bins = 20, color = 'orange', edgecolor = 'brown', kde = True)

    # Calculate mean, median, mode
    mean_value = diabetes[feature].mean()
    median_value = diabetes[feature].median()
    mode_value = diabetes[feature].mode()[0]  # mode() returns a Series, take the first mode

    # Annotate mean, median, and mode on the plot
    ax.axvline(mean_value, color='r', label=f'Mean: {mean_value:.2f}')
    ax.axvline(median_value, color='k', label=f'Median: {median_value:.2f}')
    ax.axvline(mode_value, color='b', label=f'Mode: {mode_value:.2f}')

# Adjust spacing between subplots
plt.suptitle('Histograms of Features with Mean, Median, Mode', fontsize = 20)
plt.subplots_adjust(hspace=0.2, wspace=0.2)

plt.show()
```

```python
columns_to_replace = ['Glucose', 'Blood pressure', 'Skin thickness', 'Insulin', 'Body mass index']

for col in columns_to_replace:
    median_value = diabetes[col].median()
    diabetes[col] = diabetes[col].replace(0, median_value)

diabetes.eq(0).sum()
```

```python
plt.figure(figsize = (10, 10))
features = diabetes.columns[:-1]

for idx, feature in enumerate(features) :
    ax = plt.subplot(3, 3, idx + 1)
    sns.boxplot(diabetes[feature], orient = 'h', color = 'y')

    mean_value = diabetes[feature].mean()
    ax.axvline(mean_value, color='r', label=f'Mean: {mean_value:.2f}')

plt.suptitle('BoxPlot', fontsize = 20)
plt.subplots_adjust(hspace=0.2, wspace=0.2)
plt.show()
```

```python
diabetes = diabetes[diabetes['Pregnancies'] < 14]
diabetes = diabetes[(diabetes['Blood pressure'] < 120) & (diabetes['Blood pressure'] > 40)]
diabetes = diabetes[(diabetes['Skin thickness'] < 60) & (diabetes['Skin thickness'] > 8)]
diabetes = diabetes[diabetes['Insulin'] < 650]
diabetes = diabetes[diabetes['Body mass index'] < 51]
diabetes = diabetes[diabetes['Age'] < 80]
diabetes = diabetes[diabetes['Diabetes pedigree function'] < 2]
diabetes.info()
```

```python
plt.figure(figsize = (10, 10))
features = diabetes.columns[:-1]

for idx, feature in enumerate(features) :
    ax = plt.subplot(3, 3, idx + 1)
    sns.boxplot(diabetes[feature], orient = 'h', color = 'y')

    mean_value = diabetes[feature].mean()
    ax.axvline(mean_value, color='r', label=f'Mean: {mean_value:.2f}')

plt.suptitle('BoxPlot', fontsize = 20)
plt.subplots_adjust(hspace=0.2, wspace=0.2)
plt.show()
```

```python
for col in features :
   Q1 = diabetes[col].quantile(0.25)
   Q3 = diabetes[col].quantile(0.75)
   IQR = Q3 - Q1 # interquantile range

   lb = Q1 - 1.5 * IQR # LOWER BOUND
   ub = Q3 + 1.5 * IQR # UPPER BOUND

   Outliers = diabetes[(diabetes[col] < lb) | (diabetes[col] > ub)]

   print(f'Number of outliers in {col} = {len(Outliers)}')
```

```python
# feature and target variable
X = diabetes.drop(['Outcome'], axis = 1)
Y = diabetes['Outcome']
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# split data into train_test

from sklearn.model_selection import train_test_split

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size = 0.3, random_state = 143)
```

```python
# fit the logistics regression model

from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(Xtrain, Ytrain)
```

```python
# classification
Ypred = log_reg.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# find accuracy in the model
accuracy = accuracy_score(Ytest, Ypred)
print(f'Accuracy in test_data = {accuracy}\n')

report = classification_report(Ytest, Ypred)
print(report)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Ytest, Ypred)
print(cm)
```

```python
# To apply K_Means to classify very low, low, medium, high, very high level of diabetes
from sklearn.cluster import KMeans

diabetes= diabetes[diabetes['Outcome'] > 0]
X = diabetes.drop('Outcome', axis = 1)

kmeans = KMeans(n_clusters = 5)
kmeans.fit(X)

labels = kmeans.labels_
print(labels)
```
